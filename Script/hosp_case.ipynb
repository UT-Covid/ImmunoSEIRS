```python
data_date3 = pd.read_excel('date-fit.xlsx')
date_series = data_date3['date'].squeeze()
date_series_str = date_series.dt.strftime('%Y/%m/%d')
date_series_str3 = date_series.dt.strftime('%Y-%m-%d')
date_series_str2 = date_series.dt.strftime('%m/%d/%Y')
```

```python
def get_info2(XX, state, df2, b):
    listo = list(date_series_str)
    for x in listo:
        df = df2[df2['date'] == XX] 
        df = df[df['state'] == state]
        a = df[b]
        if df.size == 0:
            return float('nan')
        else:
            return a.squeeze()
```
